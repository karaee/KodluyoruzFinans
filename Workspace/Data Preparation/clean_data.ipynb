## Günleri Eşleştirme

```python

hs = {}
for b in banks: 
    liste = []
    liste_date = []
    a  = 0
    for i in haber[haber.bank==b].date:
        tarih = i + pd.Timedelta(days=1)  
        try:
            liste.append(banks_dict[b].loc[tarih,col].values)
            liste_date.append(tarih)
        except:
            try:
                liste.append(banks_dict[b].loc[tarih + pd.Timedelta(days=1),col].values)
                liste_date.append(tarih)
            except:
                try:
                    liste.append(banks_dict[b].loc[tarih + pd.Timedelta(days=2),col].values)
                    liste_date.append(tarih)
                except:
                    liste.append([np.nan,np.nan,np.nan,np.nan])
                    liste_date.append(np.nan)
                    a +=1
    #                 try:
    #                     liste.append(banks_dict["akbank"].loc[tarih + pd.Timedelta(days=3),col].values)
    #                 except:
    #                     try:
    #                         liste.append(banks_dict["akbank"].loc[tarih + pd.Timedelta(days=4),col].values)
    #                     except:
    #                         liste.append(np.nan)
    hs[b] = [liste,liste_date]
    print(a)
```

## Haber ve Günleri Eşleştirme

```python
result = pd.DataFrame()
for b in banks:
    ba = haber[haber.bank==b].reset_index(drop=True)
    data = pd.DataFrame(data=hs[b][0],columns=col)
    tarih = pd.DataFrame(data=hs[b][1],columns=["next_date"])
    conct = pd.concat([ba,tarih,data],axis=1).reset_index(drop=True)
    result = pd.concat([result,conct],axis=0).reset_index(drop=True)
```

```python
result["Fark %"]=result["Fark %"].str.replace("%","")
result["Fark %"]=result["Fark %"].str.replace(",",".")
result["Fark %"]=result["Fark %"].astype("float")
```

```python
def classify_std(fark,std):
    if fark >= std:
        return 1
    elif fark <= -std:
        return -1
    elif -std< fark < std:
        return 0
    
def classify_mn(fark,mn1,mn2):
    if fark >= mn1:
        return 1
    elif fark <= mn2:
        return -1
    elif mn2< fark < mn1:
        return 0
    
def classify_iqr(fark,iqre):
    if fark >= iqre/2:
        return 1
    elif fark <= -iqre/2:
        return -1
    elif -iqre/2< fark < iqre/2:
        return 0
```

```python
ls_std = []
ls_mn = []
ls_iqr = []

std_all = result.groupby("bank")["Fark %"].std()
mn1_all = result.groupby(["bank",result["Fark %"] > 0]).median()["Fark %"]
mn2_all = result.groupby(["bank",result["Fark %"] > 0]).median()["Fark %"]
iqr = result.groupby("bank").quantile(0.75) - result.groupby("bank").quantile(0.25)

for b in banks:
    std = std_all.loc[b].item()
    mn1 = mn1_all.loc[b].loc[True].item() 
    mn2 = mn2_all.loc[b].loc[False].item() 
    iqre = iqr["Fark %"][b]
    data = result[result.bank==b]
    
    for d in data.values:
        fark = d[-1]
        ls_std.append(classify_std(fark,std))
        ls_mn.append(classify_mn(fark,mn1,mn2))
        ls_iqr.append(classify_iqr(fark,iqre))
#     data = result[result.bank==b]
#     result[result.bank==b]["label_std"] = data.apply(lambda x:classify_std(x["Fark %"],std),axis=1)
#     result[result.bank==b]["label_median"] = data.apply(lambda x:classify_mn(x["Fark %"],mn1,mn2),axis=1)
```
